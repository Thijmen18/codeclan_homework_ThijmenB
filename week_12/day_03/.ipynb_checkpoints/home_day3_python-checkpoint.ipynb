Q1 - Read in all 4 credit card transaction datasets and clean up the column names.

```python
import pandas as pd
# conda install pyjanitor -c conda-forge
from janitor import clean_names
```

```python
cardbase = pd.read_csv("Data/CardBase.csv").clean_names()
```

```python
customerbase = pd.read_csv("Data/CustomerBase.csv").clean_names()
fraudbase = pd.read_csv("Data/FraudBase.csv").clean_names()
transactionbase = pd.read_csv("Data/TransactionBase.csv").clean_names()
```

```python
cardbase.columns
```

Q2 - Join the data containing card details and customer details by customer id, so that all records of card details and any matching records in customer details are kept.

```python
cardbase.merge(customerbase, how = "left", on = "cust_id")
```

Q3 - Join the data containing fraud details with transaction details so all rows of both tables are kept. What does the resulting row number tell you?

```python
transactionbase.merge(fraudbase, how = "outer", on = "transaction_id")
# Fraud flag, indicates if transaction seems a fraud
```

Q4 -  Perform a filtering join between cards and transactions so that only the card details for cards that have made at least one transaction are preserved.

- step 1: join everything
- step 2: for the card number we add a column with number of transactions in the list
- step 3: we keep only >1 transaction

```python
# nr 1
transaction_join = cardbase.merge(transactionbase, how = "outer", left_on = "card_number", right_on = "credit_card_id")
```

```python
transaction_join.head()
```

```python
# nr 2_a
nr_transactions = transaction_join.groupby("card_number").agg(nr_transactions = ('card_number', 'count'))
```

```python
nr_transactions
```

```python
# nr 2_b
transaction_join2 = transaction_join.merge(nr_transactions, how = "left", on = "card_number")
```

```python
transaction_join2
```

```python
# filter and make a final join
final_join_transactions = transaction_join2.loc[(transaction_join2.nr_transactions >= 1)].copy()
```

```python
final_join_transactions.sort_values("nr_transactions")
```
